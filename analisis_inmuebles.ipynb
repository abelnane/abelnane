**Como separar dos valores en una columna, diviendola en 2 columnas para hacer esto más legible y que se pueda entender de una mejor forma**

```python
valor = inmuebles.Valor.str.split(expand=True)
inmuebles['Moneda'] = valor[0]
inmuebles['Precio'] = valor[1]
inmuebles.sample(5)
```

```python
inmuebles.info()
```

Quitamos los '.' de algunas columnas para transformarlas

```python
inmuebles['Precio'] = inmuebles['Precio'].str.replace('.','', regex=True)
inmuebles[['Precio', 'Barrio']]
```

Conversión de datos de str a float


```python
inmuebles['inmuebles_millon'] = inmuebles.Precio.astype('float')/1000000
inmuebles.info()
```

```python
inmuebles.describe()
```

Detectamos ciertas "anomalías" para lo cual debemos realizar un "Tratamiento de datos"

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(data=inmuebles, x='inmuebles_millon', y='Tipo', hue='Tipo', palette='Set1', s=100)
plt.title('Gráfico del valor de los inmuebles')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend(title='Tipos', loc='best')
plt.show()
```
